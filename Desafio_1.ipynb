```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('top_canciones_latinoamerica.csv', dtype=str)
```

```python
data.columns
```

```python
data['Popularidad'] = pd.to_numeric(data['Popularidad'], errors='coerce')

popularidad_artista_pais = data.groupby(['País', 'Artista(s)'])['Popularidad'].mean()

artista_mas_popular_por_pais = popularidad_artista_pais.reset_index().sort_values(by=['País', 'Popularidad'], ascending=False).drop_duplicates(subset=['País'])

```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='País', y='Popularidad', data=artista_mas_popular_por_pais, hue='Artista(s)', dodge=False)
plt.title('Artista Más Popular por País para Campañas de Marketing')
plt.xlabel('País')
plt.ylabel('Popularidad Promedio')
#El siguiente codigo rota el nombre de los paises 45° para una mejor vista
plt.xticks(rotation=45)
plt.show()

```
