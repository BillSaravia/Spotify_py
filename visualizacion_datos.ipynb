```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('top_3_latinoamerica.csv')
```

Configuración de estilo de seabron

```python
sns.set(style = "whitegrid")
```

Grafico de distribucion de popularidad

```python
plt.figure(figsize=(10,6))
sns.histplot(data['Popularidad'], kde= True, bins = 30, color = "skyblue" )
plt.title("Distribucion de popularidad de las canciones")
plt.xlabel("Popularidad")
plt.ylabel("Frecuencia")
plt.show()
```

Top 3 canciones con mayor popularidad en latam


```python
top_3 = data.sort_values(by="Popularidad", ascending=False)

# Eliminamos duplicados para que no se repita la misma cancion en caso de ser la que tiene mas popularidad en varios paises
top_3 = top_3.drop_duplicates(subset="Nombre de la cación")

top_3 = top_3.head(3)


```

```python
sns.set(style="whitegrid")
plt.figure(figsize = (10,6))

bar_plot = sns.barplot(x="Popularidad", y="Nombre de la cación", data=top_3, palette = "viridis")
plt.xlabel("Popularidad")
plt.ylabel("Nombre de la canción")
plt.title("Top 3 canciones mas populares en latinoamérica")
plt.show()
```

```python
popularidad_por_pais = data.groupby("País")["Popularidad"].mean().reset_index()

popularidad_por_pais = popularidad_por_pais.sort_values(by="Popularidad", ascending=False)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
bar_plot = sns.barplot(x="Popularidad", y="País", data=popularidad_por_pais, palette="viridis", order = popularidad_por_pais['País'])

plt.xlabel("Popularidad")
plt.ylabel("País")
plt.title("Popularidad promedio de las canciones por país en latinoamérica")
plt.show()
```

```python
artista_conteo = data['Artista(s)'].value_counts().reset_index()
artista_conteo.columns = ['Artista', 'Número de Canciones']

top_10_artistas = artista_conteo.head(10)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))

bar_plot = sns.barplot(x="Número de Canciones", y="Artista", data=top_10_artistas, palette="viridis")
```
